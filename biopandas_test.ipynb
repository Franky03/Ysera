```python
import pandas as pd
import numpy as np
from biopandas.pdb import PandasPdb
```

```python
arquivo = './temp/file_21.pdb'
ppdb = PandasPdb()
ppdb.read_pdb(arquivo)
ppdb.df.keys()
```

```python
atom = ppdb.df['ATOM']
hetatm = ppdb.df['HETATM']
total = pd.concat([atom, hetatm], axis=0, sort=False)
total
```

```python
total = total.drop(['atom_number','b_factor','alt_loc','line_idx',\
                    'occupancy', 'element_symbol','charge','insertion',\
                    'segment_id','blank_1','blank_2', 'blank_3', 'blank_4'], axis=1)
total.columns
```
